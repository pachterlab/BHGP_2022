```python
def makeDF(fns, met_labels):
    metrics = []
    for fn in fns:
        acc = fn.split("/")[2]
        with open(fn, "r") as f:
            m = json.load(f)
            d = {k: m.pop(k, None) for k in met_labels}

            df = pd.DataFrame(m).T.reset_index()
            df["ds"] = acc
            for ml in met_labels:
                df[ml] = d[ml]
            metrics.append(df)
    df = pd.concat(metrics)
    return df
```

```python
fns = [os.path.join("synchromesh/data/", i, f"{i}_metrics.json") for i in observations]

met_labels = [
    "ncells",
    "ngenes",
    "density",
    "total_count",
    "avg_per_cell",
    "avg_per_gene",
    "min_cell",
    "max_cell",
    "nvals",
    "overdispersion",
]

labels = [
    "raw",
    "PF",
    "sqrt",
    "log1p",
    "log1pCP10k",
    "log1pCPM",
    "scalelog1pCP10k",
    "sctransform",
    "log1pPF",
    "PFlog1pPF",
]

nan = np.array(
    [
        "ERX2756731",
        "GSM3576415",
        "ERS4228665",
        "GSM3731527",
        "GSM3576425",
        "GSM3396166",
        "GSM3148578",
        "GSM4037323",
        "GSM3576416",
        "CRX102292",
        "GSM3576398",
        "GSM3576418",
        "CRX118014",
        "GSM3576429",
        "GSM3576410",
    ],
    dtype=object,
)

df = makeDF(fns, met_labels)

# drop nans
df = df[~df["ds"].isin(nan)]

# compute 1-spearman
df["r_mono"] = 1 - np.abs(df["r_mono"])

# compute delta max min cell
df["delta_cell"] = df["max_cell"] - df["min_cell"]

mall = df.copy()

# filter by avg_per_cell
min_avg_per_cell = (
    mall.query("index == 'raw'")
    .query("ds == 'angelidis_2019'")["avg_per_cell"]
    .values[0]
)
passing_ds = np.unique(mall.query(f"avg_per_cell >= {min_avg_per_cell}")["ds"])

mall.query("index == 'raw'").query("ds == 'angelidis_2019'")
```

```python
np.random.seed(19680801)

pts = np.random.rand(30) * 0.2
# Now let's make two outlier points which are far away from everything.
pts[[3, 14]] += 0.8

# If we were to simply plot pts, we'd lose most of the interesting
# details due to the outliers. So let's 'break' or 'cut-out' the y-axis
# into two portions - use the top (ax1) for the outliers, and the bottom
# (ax2) for the details of the majority of our data
fig, (ax1, ax2) = plt.subplots(2, 1, sharey=True)
fig.subplots_adjust(wspace=0.05)  # adjust space between axes

# plot the same data on both axes
ax1.plot(pts)
ax2.plot(pts)

# zoom-in / limit the view to different portions of the data
ax1.set_ylim(0.78, 1.0)  # outliers only
ax2.set_ylim(0, 0.22)  # most of the data

# hide the spines between ax and ax2
ax1.spines.bottom.set_visible(False)
ax2.spines.top.set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

# Now, let's turn towards the cut-out slanted lines.
# We create line objects in axes coordinates, in which (0,0), (0,1),
# (1,0), and (1,1) are the four corners of the axes.
# The slanted lines themselves are markers at those locations, such that the
# lines keep their angle and position, independent of the axes size or scale
# Finally, we need to disable clipping.

d = 0.5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(
    marker=[(-1, -d), (1, d)],
    markersize=12,
    linestyle="none",
    color="k",
    mec="k",
    mew=1,
    clip_on=False,
)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)


plt.show()
```
