```python
def makeDF(fns, met_labels):
    metrics = []
    for fn in fns:
        acc = fn.split("/")[2]
        with open(fn, "r") as f:
            m = json.load(f)
            d = {k: m.pop(k, None) for k in met_labels}

            df = pd.DataFrame(m).T.reset_index()
            df["ds"] = acc
            for ml in met_labels:
                df[ml] = d[ml]
            metrics.append(df)
    df = pd.concat(metrics)
    return df
```

```python
fns = [os.path.join("synchromesh/data/", i, f"all_genes/{i}_all_genes_metrics.json") for i in observations]

met_labels = [
    "ncells",
    "ngenes",
    "density",
    "total_counts",
    "avg_per_cell",
    "avg_per_gene",
    "min_cell",
    "max_cell",
    "nvals",
    "overdispersion",
]

labels = [
    "raw",
    "PF",
    "sqrt",
    "log1p",
    "log1pCP10k",
    "log1pCPM",
    "scalelog1pCP10k",
    "sctransform",
    "log1pPF",
    "PFlog1pPF",
]

nan = np.array(['ERS4228665',
 'GSM3576415',
 'GSM3576416',
 'GSM3576418',
 'CRX102286',
 'CRX102292',
 'GSM3576425',
 'GSM3731527',
 'GSM3396166',
 'GSM3148578',
 'GSM3576429',
 'CRX118014',
 'ERX2756731',
 'GSM3576398',
 'GSM4037323',
 'GSM3576410'])

df = makeDF(fns, met_labels)

# drop nans
df = df[~df["ds"].isin(nan)]

# compute 1-spearman
df["r_mono"] = 1 - np.abs(df["r_mono"])

# compute delta max min cell
df["delta_cell"] = df["max_cell"] - df["min_cell"]

mall = df.copy()
# filter by avg_per_cell
min_avg_per_cell = (
    mall.query("index == 'raw'")
    .query("ds == 'angelidis_2019'")["avg_per_cell"]
    .values[0]
)
passing_ds = np.unique(mall.query(f"avg_per_cell >= {min_avg_per_cell}")["ds"])

mall.query("index == 'raw'").query("ds == 'angelidis_2019'")
```
